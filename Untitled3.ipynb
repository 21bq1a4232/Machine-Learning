# **Learning Machine Learning**


## **Loading Dataset**
random dataset available on github.

```python
import pandas as pd
dt=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
dt
```

```python
y=dt['logS']

X=dt.drop('logS',axis=1)
X
```

```python
from sklearn.model_selection import train_test_split

X_train ,X_test, y_train ,y_test = train_test_split(X,y,test_size=0.2, random_state=100)
X_train
```

```python
y_test
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_trainlr = lr.predict(X_train)
y_testlr = lr.predict(X_test)
```

```python
y_trainlr
```

```python
from sklearn.metrics import mean_squared_error,r2_score

ytrainmse = mean_squared_error(y_train,y_trainlr)
ytrainr2 = r2_score(y_train,y_trainlr)

ytestmse = mean_squared_error(y_test,y_testlr)
ytestr2 = r2_score(y_test,y_testlr)
```

```python
print("the trained mse is ",ytrainmse)
print("the trained r2 is ",ytrainr2)
print("the trained mse is ",ytestmse)
print("the trained mse is ",ytestr2)
```

```python
lr_results = pd.DataFrame(['Linear Regression',ytrainmse,ytrainr2,ytestmse,ytestr2]).transpose()
lr_results.columns=['Method','Trained MSE','Trained R2','Test MSE','Test R2']
lr_results
```

# Training the Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(X_train,y_train)
```

```python
apred = rf.predict(X_train)
bpred = rf.predict(X_test)
```

```python
atrainmse = mean_squared_error(y_train,apred)
atrainr2 = r2_score(y_train,apred)

btestmse = mean_squared_error(y_test,bpred)
btestr2 = r2_score(y_test,bpred)
```

```python
rf_results = pd.DataFrame(['Random Forest',atrainmse,atrainr2,btestmse,btestr2]).transpose()
rf_results.columns=['Method','Trained MSE','Trained R2','Test MSE','Test R2']
rf_results
```

```python
concat=pd.concat([lr_results,rf_results],axis=0)
concat
```
